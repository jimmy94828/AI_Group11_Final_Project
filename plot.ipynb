## Definition

#### Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### value type

```python
value_types = ['RMSE', 'MAE', 'R2']
```

## Plot: Model Comparison (4 Models)

#### baseline_all + group_all + street_all + global_all

```python
files = ['./model_baseline/linear_regression_result_all.csv', 
        './model_group/LSTM_group_results_all.csv',
        './model_street/LSTM_street_results_all.csv',
        './model_global/LSTM_global_results_all.csv']
labels = ['baseline', 'LSTM_group', 'LSTM_street', 'LSTM_global']
colors = ['blue', 'orange', 'green','red']

for value_type in value_types:
    plt.figure(figsize = (24, 9))

    for file, label, color in zip(files, labels, colors):
        df = pd.read_csv(file)
        df = df.sort_values('street')
        plt.plot(df['street'], df[value_type], label = label, color = color)

    plt.xlabel('Street')
    plt.ylabel(value_type)
    plt.title(f'{value_type} per Street with all features') 
    plt.xticks(rotation = 90)  
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'plot/4models_all_{value_type}.png') ##### TODO: modify file name
```
